# 다중회귀 = 보스톤 주택 가격

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

- 산점도와 선형회귀선: seaborn.regplot()

```python
features = df.columns[1:-1]
print(features)
```

```python
sns.regplot(x="CRIM", y="PRICE", data=df)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(16,12))
for i, feature in enumerate(features):
    row, col = i//4, i%4
    sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

- 어떤 feature가 가격과 상관관계가 높을까

```python
from sklearn.linear_model import LinearRegression
```

```python
y = boston.target
X = df['CRIM'].values.reshape(-1,1)
lr = LinearRegression()
lr.fit(X, y)
lr.score(X,y)
```

```python
for feature in features:
    y= boston.target
    X = df[feature].values.reshape(-1,1)
    lr = LinearRegression()
    lr.fit(X, y)
    r2 = lr.score(X, y)
    print(f'{feature}:\t{r2:4f}')
```

- 다중 선형회귀

```python
X = boston.data
y = boston.target
lr = LinearRegression()
lr.fit(X, y)
```

```python
# 다중선형회귀에서는 큰 의미 없음
lr.score(X, y)
```

```python
# weight

lr.coef_
```

```python
# bias
lr.intercept_
```

- 테스트 데이터 

```python
boston.data[0]
```

```python
# 기울기와 절편을 이용한 예측값
np.dot(lr.coef_, boston.data[0]) + lr.intercept_
```

```python
# 실제값
boston.target[0]
```

```python
# 5개 데이터에 대한 예측값 및 실제값
for i in range(5):
    pred = np.dot(lr.coef_, boston.data[i]) + lr.intercept_
    print(f'예측값: {pred:.2f}, 실제값: {boston.target[i]}')
```
